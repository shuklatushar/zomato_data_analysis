```python
df_country_code=pd.read_excel('CountryCode.xlsx')
```

```python
df_country_code.head()
```

```python
final_df=pd.merge(df,df_country_code,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
ind=final_df.Country.value_counts().index
```

```python
value=final_df.Country.value_counts().values
```

```python
plt.pie(value[:3],labels=ind[:3],autopct='%1.2f%%')
```

Observation: most of the transaction and records are from india(94%) and then US and UK

```python
final_df
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating counts'})
```

```python
ratings
```

Observation:
1. for rating between 4.5 to 4.9 the food was Excellent
2. for rating between 4.0 to 4.4 the food was Very Good
3. for rating between 3.5 to 3.9 the food was Good
4. for rating between 2.5 to 3.4 the food was Average
5. for rating between 1.8 to 2.4 the food was Poor


```python

sns.barplot(x=ratings['Aggregate rating'],y=ratings['Rating counts'],hue=ratings['Rating color'],palette=['white','red','orange','yellow','green','green'])
```

```python
#finding country names who gave ratings 0
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().rename(columns={0:'0 rating count'})
```

```python
#find out which currency is used by which country
final_df.groupby(['Country','Currency']).size().reset_index()
```

```python
#which countries have online delivery option
final_df[final_df['Has Online delivery']=='Yes'].Country.value_counts()
```

```python
#create a pie chart with city distribution
city_value=final_df.City.value_counts().values
city_name=final_df.City.value_counts().index
plt.pie(city_value[:5],labels=city_name[:5],autopct='%1.2f%%')
```
